## 摔手机测试
1. 第一层为实际的第二层
2. 1/2/3部手机供测试

```python
import sys
import numpy as np
n = input('输入测试的层数:')
n = int(n)
f1 = np.arange(n+1)
f2 = np.arange(n+1)
f3 = np.arange(n+1)
# 一部手机
for i in range(0,n+1):
    f1[i] = i
f2[1] = 1
# 两部手机
for i in range(2,n+1):
    ans = sys.maxsize
    for j in range(1,i):
        #1 运气好手机没坏 2 运气差:手机坏了
        _max = 1 + max(f2[i-j],f1[j-1])
        ans = min(ans,_max)
    f2[i] = ans
# 三部手机
for i in range(2,n+1):
    ans = sys.maxsize
    for j in range(1,i):
        #1 运气好手机没坏 2 运气差:手机坏了
        _max = 1 + max(f3[i-j],f2[j-1])
        ans = min(ans,_max)
    f3[i] = ans
print(f3[n])
```
